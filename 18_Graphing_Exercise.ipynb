```python
sns.barplot(x = 'Category', y = 'Millions of Dollars', data = df)
sns.set(rc={'figure.figsize':(50,25)})

plt.xticks(rotation=90, fontsize = 30)
plt.yticks(fontsize = 30)
plt.title('Items sold from 1972 - 2021', fontsize = 30)
plt.xlabel('Category of Purchases', fontsize = 30)
plt.ylabel('Millions of Dollars', fontsize = 40, labelpad = 50)
```

```python
sns.barplot(x = 'Sub-Category', y = 'Millions of Dollars', data = df)
sns.set(rc={'figure.figsize':(30,15)})

plt.xticks(rotation=90, fontsize = 30)
plt.yticks(fontsize = 30)
plt.title('Items sold from 1972 - 2021', fontsize = 40)
plt.xlabel('Sub-Category of Purchases', fontsize = 30)
plt.ylabel('Millions of Dollars', fontsize = 40, labelpad = 50)

plt.savefig('Graph.png', dpi = 1200)
```
