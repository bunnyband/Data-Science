```python
plt.style.use('seaborn')

df['Percentage of Dog Owners'].hist(bins = 7, edgecolor='black', label = 'Dog');
df['Percentage of Cat Owners'].hist(bins = 15, edgecolor='black', alpha = 0.7, color = 'orange', label = 'Cat');

plt.xticks(rotation = 45)
plt.ticklabel_format(useOffset=False, style='plain')
plt.xlabel('Smarts');
plt.ylabel('Probability');
plt.legend()

```

4. Create two boxplots on the same plot: one of the mean number of cats per household and the other of the mean number of dogs. An example of what this might look like is given below:

boxplots.png

```python
dogs = df['Mean Number of Dogs per household']
cats = df['Mean Number of Cats']

box1 = plt.boxplot([dogs, cats], labels = ['dogs', 'cats'], 
            notch = True, widths = .7, 
            patch_artist=True, medianprops = dict(linestyle='-', linewidth=2, color='Yellow'), boxprops = dict(linestyle='--', linewidth=2, color='Black', facecolor = 'blue', alpha = .4));

plt.setp(box1["boxes"][0], facecolor='red');
plt.setp(box1["boxes"][1], facecolor='green');
```
